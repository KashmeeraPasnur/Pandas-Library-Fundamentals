### Converters : Converters allows you to parse your input data to convert it to a desired dtype using a conversion function

#### Below first we need to assign the header, rename the columns also we see there is a column speices which has three values 0,1,2we need to assign the values as 0 = Setosa,1=Verginica,2=Versicolor

#### We need to create a function for assigning the valuesbefore reading the data


```python
def  convert_species(x):
    if int(x)==0:
        return 'setosa'
    elif int(x)==1:
        return 'verginica'
    else:
        return 'versicolor'
```

#### Applying the perticular function
#### Below we are first renaming the column names and then using the function
#### and in the function we need to use the name of the column which is renamed only

```python
iris = pd.read_csv("E:Dataset/Data/iris.csv",header=0,
                   names=['sepal_length','sepal_width','petal_length','petal_width','species'],
                  converters={"species":convert_species})
iris
```

####  Below we see in model column there are three distinct values SEL,SES and SE
#### So we need to assign common value to SE and SES and another value to SEL
#### There is an alternate method for "if else" method for creating a function, we can use lambda function aswell
#### Lambda function needs not to be created before reading the data but it can be created while reading the data

```python
cars = pd.read_csv("E:/Dataset/Data/usedcars.csv",
                  converters={'model':lambda x:
                              'Normal' if x in ('SE','SES')
                             else 'Luxary'})
cars
```
