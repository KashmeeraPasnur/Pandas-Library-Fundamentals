#### Drop the Last 2 columns and last 114 rows from df 
#### df.dropna() : Drops the rows /Columns with Na Values
#### axis(0/1)
#### axis=0 stands for rows
#### axis=1 stands for columns
#### how='any' :used when only some Na values are there
####  how='all' :used when there are all Na values in rows and Columns
#### inplace=True: used when you want to apply the changes
#### inplace=False : it doesnt effect the dataset

```python
# Here all the dataset in all rows is dropped
Air.dropna(axis=0,how='any').shape
```

```python
# Here all the dataset in all the Columns is dropped
Air.dropna(axis=1,how='any').shape
```

#### Dropping only Rows

```python
Air.dropna(axis=0,how='all')
```

```python
Air.dropna(axis=0,how='all').shape
```

#### Dropping only Columns

```python
Air.dropna(axis=1,how='all')
```

```python
Air.dropna(axis=1,how='all').shape
```

```python
Air.shape
```

#### Dropping NaN rows completly from DataFrame

```python
Air.dropna(axis=0,how='all',inplace=True)
```

### Replacing the NaN Values

#### Replacing of the Value should be done with the Average Value, from the values besides them
####  Replacing of the Value should not be done by 0 directly
#### Suppose we need to replace value of the data which is in between 50 and 52
#### e.g 50,NaN,52 =>Find average of 50 and 52 => (50+52)/2= 51 should be the assigned

#### Replacing can also be done by the forward fill: ffill or backward fill or bfill

#### fillna : Used for filling the NaN values
