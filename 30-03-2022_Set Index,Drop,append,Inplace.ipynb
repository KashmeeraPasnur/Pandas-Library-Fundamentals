#### If we want to add the index's after loading the data
#### Also we want to remove the index's after loading the data

#### set_index() : Is used to add the index
#### reset_index : Is used to remove the index

#### drop =True/False
#### If it is true,Column of which index should be maintained as a Data Column or not 
#### It deletes the column to be used as the new index

####  append = True/False
##### append is used to help us to override or append the new Index

#### inplace = True/False
##### If we dont want to make changes to the dataset then we can set inplace=False
##### Whatever the changes we have applied should be reflected in the data or not 

```python
# Here we can see year column is in the row index as well as a column Name
cars.set_index(keys='price',drop=True,append=True)
```

```python
#Here the year column has been dropped when it is True
cars.set_index(keys='year',drop=True,append=False)
```

```python
#Here the year column has been dropped when it is True
cars.set_index(keys='price',drop=True,append=True)
```

```python
# After using inplace, conditions will be applied to the dataset
cars.set_index(keys='model',drop=False,append=False,inplace=True)
```
