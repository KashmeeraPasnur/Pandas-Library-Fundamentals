### Reading fwf file format

#### fwf file means : fixed width file
#### There is a file format which is not seperated with any somma,semicolon or spaces
#### So in this case it is called fwf which has same width of all the lines
####  widths is used to give the count of each word,names is used to assign the column names to the data

```python
pd.read_fwf("E:/Dataset/Data/employee.txt",
           header=None,widths=[4,4,4,2],names=['Empno','Empname','salary','deptno'] )
```
